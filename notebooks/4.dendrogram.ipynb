```python
from sklearn.cluster import AgglomerativeClustering
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
import scipy.cluster.hierarchy as shc
```

```python
df_pca = pd.read_csv("data/pca_dataset.csv", index_col=0)
```

```python
print(f"Датасет для кластеризации готов: {df_pca.shape}")
plt.figure(figsize=(25, 10))
plt.title("Дендрограмма развития стран (Method: Ward)", fontsize=16)
dend = shc.dendrogram(shc.linkage(df_pca, method='ward'))
plt.ylabel('Евклидово расстояние')
plt.xlabel('Страны (индексы)')
plt.show()
```

```python
# анализ стран при 3 компонентах
agg_model = AgglomerativeClustering(n_clusters=3, linkage='ward')
agg_labels = agg_model.fit_predict(df_pca)
df_hierarchical = df_pca.copy()
df_hierarchical['Hierarchical_Cluster'] = agg_labels
for cluster_id in range(4):
    countries = df_hierarchical[df_hierarchical['Hierarchical_Cluster'] == cluster_id].index.tolist()
    print(f"\n--- Cluster {cluster_id} (Всего стран: {len(countries)}) ---")
    print(", ".join(countries))
```

Cluster 0: Развивающиеся страны и страны с низким доходом
Этот самый большой кластер (135 стран) объединяет преимущественно развивающиеся экономики, включая большинство стран Африки (AFE, AFG, AFW, AGO, BEN, BFA), Южной Азии (BGD, BTN), Латинской Америки (ARG, BOL) и Китай (CHN). В группу входят наименее развитые страны (AFG, BDI, CAF, COD, YEM, ZMB, ZWE) и страны со средним уровнем дохода, которые объединяет относительно низкий ВВП на душу населения, развивающаяся инфраструктура и схожие социально-экономические показатели. Также присутствуют агрегированные коды регионов (AFE - Африка Восточная, AFW - Африка Западная, ARB - Арабские страны, CEB - Центральная Европа и Балтия).​

Cluster 1: Развитые экономики и страны с высоким доходом
Кластер из 82 стран включает развитые страны и успешные emerging markets: США, Канаду, Западную Европу (AUT, BEL, CHE, DEU), Австралию, а также крупные развивающиеся экономики с высоким потенциалом (BRA, CHL, COL, CHN). Сюда входят страны ОЭCD, нефтедобывающие монархии Персидского залива (ARE, BHR, KWT, SAU), постсоветские страны среднего уровня развития (ARM, AZE, BLR, UKR) и индустриальные страны Азии (JPN, KOR, SGP). Код WLD (World) указывает, что это странs с показателями близкими к среднемировым или выше.​

Cluster 2: Малые островные государства и офшорные территории
Этот кластер из 49 стран состоит преимущественно из малых островных развивающихся государств (МО РГ), офшорных финансовых центров и территорий с особым статусом: Карибские острова (ATG, BRB, BHS, BLZ, DMA, GRD), тихоокеанские территории (ASM, FSM, GUM, PLW, VUT, WSM), европейские офшоры (AND, BMU, CYM, GIB, IMN) и Южная Африка (ZAF). Эти территории объединяет небольшой размер, островное положение, зависимость от туризма или финансовых услуг и уникальные экономические модели, отличающие их от континентальных стран.​



```python
# анализ стран при 4 компонентах
agg_model = AgglomerativeClustering(n_clusters=4, linkage='ward')
agg_labels = agg_model.fit_predict(df_pca)
df_hierarchical = df_pca.copy()
df_hierarchical['Hierarchical_Cluster'] = agg_labels
for cluster_id in range(4):
    countries = df_hierarchical[df_hierarchical['Hierarchical_Cluster'] == cluster_id].index.tolist()
    print(f"\n--- Cluster {cluster_id} (Всего стран: {len(countries)}) ---")
    print(", ".join(countries))
```

Cluster 0: Наименее развитые страны (НРС)
Этот кластер из 46 стран представляет наименее развитые страны (LDC) согласно классификации ООН. Сюда входят беднейшие страны Африки (AFG, BDI, BEN, BFA, CAF, COD, ETH, GIN, GMB, TGO, UGA, ZWE), азиатские НРС (LAO, MMR, NPL, TLS, YEM) и единственная карибская страна этой категории - Гаити (HTI). Эти страны характеризуются низким ВВП на душу населения (менее $1,135), слабым человеческим капиталом и высокой экономической уязвимостью. В группе также присутствуют отдельные страны с низким доходом (EGY, KEN, NGA, PAK), не входящие формально в список НРС, но имеющие схожие показатели.​

82 страны включают развитые экономики (USA, CAN, AUS, западноевропейские страны), нефтедобывающие монархии Персидского залива (ARE, BHR, KWT, SAU), успешные emerging markets (BRA, CHL, COL, TUR) и постсоветские страны среднего уровня (ARM, AZE, BLR, UKR). Cluster 2: Малые островные государства и офшоры
Без изменений - 49 малых островных развивающихся государств, офшорных центров и территорий с особым статусом (карибские и тихоокеанские острова, европейские офшоры, Южная Африка).​

Cluster 3: Страны со средним уровнем дохода и региональные агрегаты
Новый кластер из 89 стран объединяет развивающиеся экономики среднего уровня дохода (lower-middle и upper-middle income): Китай (CHN), латиноамериканские страны (BOL, DOM, ECU, VEN), азиатские emerging markets (BGD, BTN, VNM, UZB), африканские страны (AGO, COG, DZA, ZMB) и многочисленные региональные агрегаты (AFE, AFW, ARB, CEB, EAP, EAS, ECA). Эти страны имеют ВВП на душу населения от $1,136 до $13,935 и активно развивающуюся экономику.​
